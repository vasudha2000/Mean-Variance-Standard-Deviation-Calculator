```python
#create dictionary value for each mean, variance etc and add 
import numpy as np
def mean(matrix):
    mean=np.mean(matrix,axis=0)
    l=[i for i in mean]
    mean1=np.mean(matrix,axis=1)
    v=[i for i in mean1]
    mean2=np.mean(matrix)
    mean_list=[]
    for i in range(1):
        mean_list.append(l)
        mean_list.append(v)
        mean_list.append(mean2)
    return mean_list

def variance(matrix):
    variance=np.var(matrix,axis=0)
    l=[i for i in variance]
    variance1=np.var(matrix,axis=1)
    v=[i for i in variance1]
    variance2=np.var(matrix)
    variance_list=[]
    for i in range(1):
        variance_list.append(l)
        variance_list.append(v)
        variance_list.append(variance2)
    return variance_list

def standard_dev(matrix):
    stan_dev=np.std(matrix,axis=0)
    l=[i for i in stan_dev]
    stan_dev1=np.std(matrix,axis=1)
    v=[i for i in stan_dev1]
    stan_dev2=np.std(matrix)
    stan_dev_list=[]
    for i in range(1):
        stan_dev_list.append(l)
        stan_dev_list.append(v)
        stan_dev_list.append(stan_dev2)
    return stan_dev_list

def maximum(matrix):
    maximum=np.max(matrix,axis=0)
    l=[i for i in maximum]
    max1=np.max(matrix,axis=1)
    v=[i for i in max1]
    max2=np.max(matrix)
    max_list=[]
    for i in range(1):
        max_list.append(l)
        max_list.append(v)
        max_list.append(max2)
    return max_list

def minimum(matrix):
    minimum=np.min(matrix,axis=0)
    l=[i for i in minimum]
    min1=np.min(matrix,axis=1)
    v=[i for i in min1]
    min2=np.min(matrix)
    min_list=[]
    for i in range(1):
        min_list.append(l)
        min_list.append(v)
        min_list.append(min2)
    return min_list

def Total(matrix):
    s=np.sum(matrix,axis=0)
    l=[i for i in s]
    sum1=np.sum(matrix,axis=1)
    v=[i for i in sum1]
    sum2=np.sum(matrix)
    sum_list=[]
    for i in range(1):
        sum_list.append(l)
        sum_list.append(v)
        sum_list.append(sum2)
    return sum_list


def calculate(a1):
    if(len(a1) != 9):
        raise ValueError("List must contain nine numbers.")
    else:
        arr=np.array(a1)
        matrix=np.reshape(a1,(3,3))
        a=mean(matrix)
        b=maximum(matrix)
        c=minimum(matrix)
        d=Total(matrix)
        e=variance(matrix)
        f=standard_dev(matrix)
        dic={}
        for i in range(len(a)):
            dic['mean']=a
            dic['variance']=e
            dic['standard deviation']=f
            dic['max']=b
            dic['min']=c
            dic['sum']=d
        print(dic)
calculate([9,1,5,3,3,3,9,0,4])


```
